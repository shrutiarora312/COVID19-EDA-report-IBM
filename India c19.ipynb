```python
#Plotting the Bar Graph 
df1 = df.head()
import matplotlib.pyplot as plt
plt.bar(df1.Date,df1.ConfirmedIndianNational,color=['y'])
plt.xlabel('Date')
plt.ylabel('Indian Cases')
plt.title('Bar Graph')
plt.show()
```

```python
#Scatter Plot 
df2 = df.head()
plt.scatter(df2['Date'],df2['ConfirmedForeignNational'])
plt.show()
```
